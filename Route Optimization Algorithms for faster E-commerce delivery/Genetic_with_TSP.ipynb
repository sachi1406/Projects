```python
import random
import numpy as np

# Define the distance matrix
# distance_matrix = np.array([[0, 10, 15, 20],
#                             [10, 0, 35, 25],
#                             [15, 35, 0, 30],
#                             [20, 25, 30, 0]])

distance_matrix = np.array(dist_matrix)

# Define the GA parameters
population_size = 1500
mutation_rate = 0.01
num_generations = 500

# Define the fitness function
def fitness(solution):
    distance = 0
    for i in range(len(solution) - 1):
        distance += distance_matrix[solution[i]][solution[i + 1]]
    distance += distance_matrix[solution[-1]][solution[0]]  # add distance from last city to first city
    return distance


# Define the initial population
def create_initial_population():
    population = []
    for i in range(population_size):
        population.append([0] + random.sample(range(1, distance_matrix.shape[0]), distance_matrix.shape[0]-1))
    return population

# Define the selection function
def selection(population):
    sorted_population = sorted(population, key=lambda x: fitness(x))
    return sorted_population[:int(0.5*len(sorted_population))]

# Define the crossover function
def crossover(parent1, parent2):
    child = [0] * len(parent1)
    start = random.randint(0, len(parent1)-2)
    end = random.randint(start+1, len(parent1)-1)
    for i in range(start, end+1):
        child[i] = parent1[i]
    j = 0
    for i in range(len(parent2)):
        if parent2[i] not in child:
            if j == start:
                j = end + 1
            child[j] = parent2[i]
            j += 1
    return child


# Define the mutation function
def mutation(solution):
    for i in range(len(solution)):
        if random.random() < mutation_rate:
            j = random.randint(1, len(solution)-1)
            solution[i], solution[j] = solution[j], solution[i]
    return solution

# Create the initial population
population = create_initial_population()

# Iterate through the generations
for generation in range(num_generations):
    # Perform selection
    selected_population = selection(population)
    
    # Perform crossover and mutation to create the new population
    new_population = []
    while len(new_population) < population_size:
        parent1, parent2 = random.sample(selected_population, 2)
        child = crossover(parent1, parent2)
        child = mutation(child)
        new_population.append(child)
    
    # Replace the old population with the new population
    population = new_population
    
    # Print the best solution in each generation
  # Print the best solution in each generation
    best_solution = min(population, key=lambda x: fitness(x))
    best_path = [0] + best_solution
    print(f"Generation {generation+1}: Best path = {best_path}, Distance = {fitness(best_solution)}")

```
